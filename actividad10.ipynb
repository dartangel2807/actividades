```python
import random
import pprint
import json
import pandas as pd

# Lista para almacenar los datos de las personas
lista_personas = []

# Función para agregar una persona automáticamente
def agregarAuto():
    id = random.choice([372611,372612,372613,372614,372615,372616,372617,372618,372619,372620])
    nombre = random.choice(["Angel", "Ricardo", "Pedro", "Kevin", "Brandon", "Elizabeth", "Jose", "Luis", "Carlos", "Jonathan","Maria","Juliette","Dulce"])
    aPaterno = random.choice(["Gonzalez", "Valenzuela", "Coronado", "Hernandez", "Vasquez", "Sanchez", "Ontiveros", "Hedilla", "Rivera", "Lopez"])
    aMaterno = random.choice(["Perez", "Fernandez", "Osuna", "Ebrard", "Meave", "Silva", "Cano", "Cortez", "Cardenas", "Suarez"])
    edad = random.randint(30, 39)
    sexo = "H" if nombre in ["Angel", "Ricardo", "Pedro", "Kevin", "Brandon", "Jose", "Luis", "Carlos", "Jonathan"] else "M"

    persona = {"id": id, "Nombre": nombre, "Ap. Paterno": aPaterno, "Ap. Materno": aMaterno, "Edad": edad, "Sexo": sexo}
    lista_personas.append(persona)

# Función para agregar una persona manualmente
def agregarManual():
    id = int(input("Escribe tu ID: "))
    nombre = input("Escribe tu(s) nombre(s): ")
    aPaterno = input("Escribe tu apellido paterno: ")
    aMaterno = input("Escribe tu apellido materno: ")
    edad = int(input("Escribe tu edad: "))
    sexo = input("Escribe tu sexo (H) o (M): ")

    persona = {"id": id, "Nombre": nombre, "Ap. Paterno": aPaterno, "Ap. Materno": aMaterno, "Edad": edad, "Sexo": sexo}
    lista_personas.append(persona)

# Función para imprimir la lista de personas usando pprint, json o pandas aleatoriamente
def imprimirLista():
    if not lista_personas:
        print("La lista de personas está vacía.")
        return

    opcion_imprimir = random.choice([pprint.pprint, json.dumps, pd.DataFrame])
    if opcion_imprimir == pd.DataFrame:
        df = pd.DataFrame(lista_personas)
        print(df)
    else:
        opcion_imprimir(lista_personas)

# Función para buscar a un trabajador por su ID
def buscarPorID():
    id_buscar = int(input("Ingresa el ID del trabajador: "))
    encontrado = False
    for persona in lista_personas:
        if persona["id"] == id_buscar:
            print("Trabajador encontrado:")
            pprint.pprint(persona)
            encontrado = True
            break
    if not encontrado:
        print("No encontré ningún trabajador con esa ID")

# Función para ordenar la lista
def ordenarPorID():
    lista_personas.sort(key=lambda x: x["id"])
    print("Lista ordenada por ID:")
    imprimirLista()

# Función para eliminar a un trabajador por ID
def eliminarPorID():
    id_eliminar = int(input("Ingresa el ID de la persona que deseas eliminar: "))

    encontrado = False
    for i, persona in enumerate(lista_personas):
        if persona["id"] == id_eliminar:
            print("Persona encontrada:")
            pprint.pprint(persona)
            confirmacion = input("¿Estás seguro de que deseas eliminar esta persona? (S/N): ")
            if confirmacion.lower() == "s":
                del lista_personas[i]
                print("Persona eliminada.")
            else:
                print("Operación cancelada.")
            encontrado = True
            break
    if not encontrado:
        print("No se encontró ninguna persona con el ID proporcionado.")

# Función para eliminar toda la lista
def eliminarTodaLista():
    confirmacion = input("¿Estás seguro de que deseas eliminar toda la lista? (S/N): ")
    if confirmacion.lower() == "s":
        lista_personas.clear()
        print("Ha sido eliminada toda la lista.")
    else:
        print("Operación cancelada.")

# Menu de opciones
try:
  while True:
      print("\nMenu de opciones:")
      print("1.- Agregar (automático)")
      print("2.- Agregar (manual)")
      print("3.- Imprimir lista")
      print("4.- Buscar {ID}")
      print("5.- Ordenar {ID}")
      print("6.- Eliminar {ID}")
      print("7.- Borrar toda la lista")
      print("0.- Salir")
      opcion = int(input("Seleccione una opción: "))
      if opcion == 1:
          agregarAuto()
      elif opcion == 2:
          agregarManual()
      elif opcion == 3:
          imprimirLista()
      elif opcion == 4:
          buscarPorID()
      elif opcion == 5:
          ordenarPorID()
      elif opcion == 6:
          eliminarPorID()
      elif opcion == 7:
          eliminarTodaLista()
      elif opcion == 0:
          break
except ValueError:
  print("Sólamente coloca números enteros como opción!")
```
